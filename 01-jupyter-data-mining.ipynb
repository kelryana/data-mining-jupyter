```python
# 🐍 Meu Primeiro Projeto de Data Mining
# Ana Kelry — Julho 2026

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("✅ Jupyter pronto para Data Mining!")
print("🐍 Python funcionando. Pacotes carregados com sucesso.")
```

```python
# 📊 Carregando o dataset Íris
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['espécie'] = iris.target_names[iris.target]

print(f"📏 Tamanho do dataset: {df.shape[0]} linhas, {df.shape[1]} colunas")
print(f"🌺 Espécies: {df['espécie'].unique()}")
df.head()
```

```python
# 📈 Estatísticas descritivas
df.describe()
```

```python
# 🎨 Gráfico de dispersão
sns.pairplot(df, hue='espécie', height=2)
plt.suptitle('Análise Exploratória — Dataset Íris', y=1.02, fontsize=14)
plt.show()
```
